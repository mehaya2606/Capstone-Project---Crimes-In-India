```python
cluster=pd.merge(district_wise,districtwise_sctot,how='inner',left_on=['STATE/UT','DISTRICT','Year'],right_on=['STATE/UT','DISTRICT','Year']).merge(districtwise_sttot,how='inner',left_on=['STATE/UT','DISTRICT','Year'],right_on=['STATE/UT','DISTRICT','Year']).merge(districtwise_childtot,how='inner',left_on=['STATE/UT','DISTRICT','Year'],right_on=['STATE/UT','DISTRICT','Year']).merge(districtwise_womentot,how='inner',left_on=['STATE/UT','DISTRICT','Year'],right_on=['STATE/UT','DISTRICT','Year'])
```

```python
cluster=cluster.groupby(['STATE/UT','DISTRICT']).sum().reset_index().drop(['Year'],axis=1)
```

```python
cluster
```

# 4.1 CLUSTERING USING THE KMEANS

```python
x=cluster.iloc[:,2:]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=3,random_state=0)
```

```python
y_kmeans=kmeans.fit_predict(x)
```

# 4.2 Create DataFrame for each cluster that shows data according to the areas.

```python
sensitive_regions=cluster[y_kmeans==2]
```

```python
sensitive_regions
```

```python
moderate_regions=cluster[y_kmeans==1]
```

```python
peaceful_regions=cluster[y_kmeans==0]
```

## What needs to be done to reduce crime.

Reducing crime in India requires a multifaceted approach. Here are some strategies that could be effective:

Crime Prevention through Environmental Design (CPTED): This approach focuses on modifying the physical environment to reduce crime opportunities.

Defensible Spaces: Creating spaces that naturally deter criminal activities through their design.

Situational Crime Prevention: Implementing measures that increase the effort and risks for committing crimes while reducing the rewards.

Community Crime Prevention: Engaging communities to take collective responsibility for preventing crime.

Routine Activities and Crime Patterns: Analyzing and altering routines and patterns to prevent crime.

Proactive Policing: Adopting strategies that focus on preventing crime before it occurs.

Expedite the Judicial Process: Ensuring swift and fair trials to deter potential criminals.

Educating Communities: Raising awareness about crime prevention and laws.

Additionally, the government can focus on preventing conditions that lead to criminality, such as poverty, lack of education, and unemployment4. It’s also crucial to have robust laws and an effective justice system to bring criminals to justice.
